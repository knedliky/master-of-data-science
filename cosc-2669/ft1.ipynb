## Data Preparation
#### Importing and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", None, "display.max_columns", None)

import warnings
warnings.filterwarnings('ignore')
```

```python
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```python
train = train.drop(columns ='ID')
test = test.drop(columns ='ID')
```

Dropping the ID column as it is redundant

```python
train.head()
```

```python
test.head()
```

```python
y = train['TARGET']
X = train.drop(columns='TARGET')
```

```python
X.describe()
```

Nothing looks awry with the data, it should be now suitable for standardisation and normalisation

#### Standardising

```python
from sklearn import preprocessing

minmax = preprocessing.MinMaxScaler().fit_transform(X)

X = pd.DataFrame(minmax, columns = X.columns)
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

## Experimentation
#### Support Vector Machine (Classification)

```python
from sklearn import svm

clf = svm.SVC()
clf = clf.fit(X_train, y_train)
```

```python
clf_predict = clf.predict(X_test)
```

#### Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, clf_predict)
```

A Support Vector Machine provides a 96% accuracy score for predicting Customer Satisfaction, which is unusually high. This may indicate that the target classes are unproportional.

#### Gaussian Naive Bayes (Classification)

```python
from sklearn.naive_bayes import GaussianNB

clf_gnb = GaussianNB()
clf_gnb = clf_gnb.fit(X_train, y_train)
```

```python
clf_gnb_predict = clf_gnb.predict(X_test)
```

#### Accuracy

```python
accuracy_score(y_test, clf_gnb_predict)
```

A Gaussian Naive Bayes Classifier achieves a 9.7% accuracy score for predicting Customer Satisfcation, much lower than a Support Vector Machine, indicating that the data is not normally distributed and there may be several outliers in the data.

```python
import tensorflow as tf
import functools
```

```python
# Converts a dataframe into a list of tf.Example protos.
def df_to_examples(df, columns=None):
    examples = []
    if columns == None:
        columns = df.columns.values.tolist()
    for index, row in df.iterrows():
        example = tf.train.Example()
        for col in columns:
            if df[col].dtype is np.dtype(np.int64):
                example.features.feature[col].int64_list.value.append(int(row[col]))
            elif df[col].dtype is np.dtype(np.float64):
                example.features.feature[col].float_list.value.append(row[col])
            elif row[col] == row[col]:
                example.features.feature[col].bytes_list.value.append(row[col].encode('utf-8'))
        examples.append(example)
    return examples
```
